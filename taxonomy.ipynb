# Taxonomy

```python
import pandas as pd
taxonomy = pd.read_csv("/kaggle/input/happywhalespeciesclassification/species.csv")
taxonomy.shape
```

```python
#taxonomy.sort_values(by = ['infraorder','family','genus','specy']).drop(columns = ['wikipedia','image','size']).set_index('specy_id')
taxonomy = taxonomy.sort_values(by = ['specy']).drop(columns = ['wikipedia','image','size']).set_index('specy_id')
```

```python
taxonomy
```

```python
df = pd.read_csv("/kaggle/input/happy-whale-and-dolphin/train.csv")
df = df.rename(columns = { 'species': 'specy_id' })
pd.merge(df, taxonomy, on = ['specy_id'], how = 'left')
```
